```python
import pandas as pd
import numpy as np
```

```python
tmdb_credits = pd.read_csv('database/tmdb/tmdb_5000_credits.csv')
tmdb_credits.head()
```

```python
tmdb_movies = pd.read_csv('database/tmdb/tmdb_5000_movies.csv')
tmdb_movies.head()
```

```python
print('1.For each movie, compute the number of cast members')
pd.Series([len(pd.DataFrame(eval(cast))) for cast in tmdb_credits.cast], index=tmdb_credits.title)
```

```python
print('2.How many movies do not have a homepage?')
len(tmdb_movies[tmdb_movies.homepage.isnull()])
```

```python
print('3.For each year, how many movies do not have a homepage?')
tmdb_movies[tmdb_movies.homepage.isnull()].groupby(tmdb_movies.release_date.str[:4]).size()
```

```python
print('4.Extract the domain of each homepage.')
pd.Series(np.array(tmdb_movies[tmdb_movies.homepage.notnull()].homepage.str.partition('//')[2].str.partition('/')[0]), index=tmdb_movies[tmdb_movies.homepage.notnull()].title)
```

```python
print('5.Extract of a set of normalized tables. That is, each entry of a normalized table must contain exactly one value (not a list or a dictionary)')

def normalize(column_to_normalize, key_column):
    new_df = pd.DataFrame()
    keys = np.array([])
    for i in range(len(column_to_normalize)):
        row_df = pd.DataFrame(eval(column_to_normalize[i]))
        new_df = pd.concat([new_df, row_df], ignore_index=True)
        keys = np.concatenate([keys, [key_column[i]] * len(row_df)])
    new_df['key'] = pd.Series(keys)
    return new_df

#normalization of tmdb_credits
tmdb_cast = normalize(tmdb_credits.cast, tmdb_credits.title)
tmdb_cast.head()
```

```python
tmdb_crew = normalize(tmdb_credits.crew, tmdb_credits.title)
tmdb_crew.head()
```

```python
#normalization of tmdb_movies
tmdb_genres = normalize(tmdb_movies.genres, tmdb_movies.title)
tmdb_genres.head()
```

```python
tmdb_keywords = normalize(tmdb_movies.keywords, tmdb_movies.title)
tmdb_keywords.head()
```

```python
tmdb_production_companies = normalize(tmdb_movies.production_companies, tmdb_movies.title)
tmdb_production_companies.head()
```

```python
tmdb_production_countries = normalize(tmdb_movies.production_countries, tmdb_movies.title)
tmdb_production_countries.head()
```

```python
tmdb_spoken_languages = normalize(tmdb_movies.spoken_languages, tmdb_movies.title)
tmdb_spoken_languages.head()
```

```python
tmdb_credits.drop(['cast', 'crew'], axis=1, inplace=True)
tmdb_movies.drop(['genres', 'keywords', 'production_companies', 'production_countries', 'spoken_languages'], axis=1, inplace=True)
```

```python
print('6.For each movie, compute the gross margin (difference between revenue and budget)')
pd.Series(np.array(tmdb_movies.revenue - tmdb_movies.budget), index=tmdb_movies.title)
```

```python
print('7.For each movie, compute the number of crew members')
tmdb_crew.groupby('key').size()
```

```python
print('8.For each movie, compute the number of directors')
tmdb_crew[tmdb_crew.job=='Director'].groupby('key').size()
```

```python
print('9.For each language, compute the number of movies where such language is spoken')
tmdb_spoken_languages.groupby('iso_639_1').size()
```

```python
print('10.For each company and each decade, compute the overall revenue')
tmdb_movies_companies = pd.merge(tmdb_movies, tmdb_production_companies, left_on='title', right_on='key')
tmdb_movies_companies.groupby(['name', tmdb_movies_companies.release_date.str[:3]+'0']).sum().revenue
```

```python
print('11.For each decade, compute the company with maximum revenue')
grouped_revenue = tmdb_movies_companies.groupby([tmdb_movies_companies.release_date.str[:3]+'0', 'name']).sum().revenue
grouped_revenue.groupby(level=0).apply(lambda x: x.idxmax()[1])
```

```python
print('12.In each year, how many movies have revenue smaller than the budget')
tmdb_movies[tmdb_movies.revenue < tmdb_movies.budget].groupby(tmdb_movies.release_date.str[:4]).size()
```
